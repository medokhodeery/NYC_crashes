```python
df=df.drop(axis=1, columns=['longitude','latitude']) #repition of the same info in location feature
```

```python
df=df.drop(axis=1 , columns=['collision_id']) #not important information for machine learning
```

```python
print(df['vehicle_type_code1'].unique()) # to reduce the number of different vechile types types 
```

```python
trucks = ['USPS TRUCK','Pick up','PICK-UP TR','Trc','Pick-up Truck', 'Tractor Truck Diesel', 'Box Truck', 'Tractor Truck Gasoline', 'Tow Truck / Wrecker', 'Beverage Truck', 'truck', 'Armored Truck', 'TRUCK', 'Truck', 'Tow Truck', 'FIRE TRUCK', 'PICK-UP TRUCK', 'BOX TRUCK', 'Fire Truck', 'box truck', 'TRUCK FLAT', 'FIRTRUCK', 'TRUCK VAN', 'FIRETRUCK', 'FDNY TRUCK', 'dump truck', 'Fire truck', 'UPS TRUCK', 'fire truck', 'FOOD TRUCK', 'MAIL TRUCK', 'USPS TRUCK', 'tow truck', 'TOW TRUCK', 'Tow truck']
df['vehicle_type_code1'] = df['vehicle_type_code1'].replace(to_replace=trucks, value='Truck')


amb= ['Ambulance', 'ambulance','EMS Ambula','AMBULANCE', 'EMS/VAN', 'AMBUL', 'Fdny ambul', 'NYC AMBULA',]
df['vehicle_type_code1'] = df['vehicle_type_code1'].replace(to_replace=amb, value='Ambulance')

sed= ['2 dr sedan', 'Sedan', '4 dr sedan']
df['vehicle_type_code1'] = df['vehicle_type_code1'].replace(to_replace=sed, value='Sedan')

van = ['van', 'Van', 'CARGO VAN', 'WORK VAN','Work van'] 
df['vehicle_type_code1'] = df['vehicle_type_code1'].replace(to_replace=van, value='Van')

print(df['vehicle_type_code1'].unique())
```

```python
df.duplicated().any()
df.drop_duplicates(inplace=True)
```

```python
df=df.dropna(subset=['location','zip_code','borough'], how='all')
```

```python
df.head(20)
```

```python
df['zip_code'] = df['zip_code'].fillna(0)  #replace NAN with zero in zip code
df['on_street_name'] = df['on_street_name'].fillna(0)
```
